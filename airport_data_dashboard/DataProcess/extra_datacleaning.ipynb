```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("final_flight_data.csv")
df.head()
```

```python
a = df['Origin'].unique()
a = pd.DataFrame(a)
a = a.rename(columns={0: 'iata_code'})
a
```

```python
df2 = pd.read_csv("all_airport_info.csv")
df2.head()
```

```python
Merged = pd.merge(a, df2, on = 'iata_code')
Merged = Merged.drop_duplicates(subset=['iata_code'], keep='first')
Merged = Merged.reset_index()
Merged.to_csv("final_airport_info.csv",index = False)
```
